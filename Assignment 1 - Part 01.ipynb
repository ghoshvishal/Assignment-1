```python
# Importing required modules

import mysql.connector as mc # To work with mysql database
import xlrd # xlrd is a module that allows Python to read data from Excel files
```

```python
# Establish mysql connection for table 1

db = mc.connect(host = 'localhost', user ='vishal18', passwd = 'Vishal@18', database ='intern')
cursor = db.cursor()
```

```python
# open the workbook and define the worksheet

excel = xlrd.open_workbook('beginner_assignment01 (1).xlsx')
sheet = excel.sheet_by_index(0)
sheet2 = excel.sheet_by_index(1)
```

```python
# Creating Table1 - product_listing
cursor = db.cursor()
cursor.execute('''CREATE TABLE product_listing
(
Product_Name VARCHAR (30),
Model_Name VARCHAR (30),
Product_Serial_No VARCHAR (30),
Group_Associated VARCHAR (30),
Product_MRP_in_Rs VARCHAR(30)
)''')
```

```python
# Create INSERT INTO sql querry for table 1

query =  '''INSERT INTO product_listing(Product_Name,Model_Name,Product_Serial_No,Group_Associated,Product_MRP_in_Rs)
Values(%s,%s,%s,%s,%s)'''
```

```python
# Create a for loop to iterate through each row in the XLS file for table 1

for r in range(1,sheet.nrows):
    Product_Name =sheet.cell(r,0).value
    Model_Name = sheet.cell(r,1).value
    Product_Serial_No = sheet.cell(r,2).value
    Group_Associated = sheet.cell(r,3).value
    Product_MRP_in_Rs = sheet.cell(r,4).value
    # Assign values from each row
    values = (Product_Name,Model_Name,Product_Serial_No,Group_Associated,Product_MRP_in_Rs)
    #Execute sql querry
    cursor.execute(query,values)
# CLose the cursor
cursor.close()

# commit the transaction
db.commit()

# close the database
db.close()

# Print results
columns = str(sheet.ncols)
print(f'{columns} : {sheet.nrows} are created in product_listing')
```

```python
# Establish mysql connection for table 2

db = mc.connect(host = 'localhost', user ='vishal18', passwd = 'Vishal@18', database ='intern')
cursor = db.cursor()
```

```python
# Create INSERT INTO sql querry for table 2

query2 =  '''INSERT INTO group_listing(Group_Name,Group_description,isActive)
Values(%s,%s,%s)'''
```

```python
# Create a for loop to iterate through each row in the XLS file for table 2

for r in range(1,sheet2.nrows):
    Group_Name =sheet2.cell(r,0).value
    Group_description = sheet2.cell(r,1).value
    isActive = sheet2.cell(r,2).value

    # Assign values from each row
    values = (Group_Name,Group_description,isActive)
    #Execute sql querry
    cursor.execute(query2,values)
# CLose the cursor
cursor2.close()

# commit the transaction
db.commit()

# close the database
db.close()

# Print results
columns = str(sheet2.ncols)
print(f'{columns} : {sheet2.nrows} are created in group_listing')
```
